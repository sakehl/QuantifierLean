```python
from typing import Tuple
```

```python
def divE(num: int, denom: int) -> Tuple[int, int]:
  rT = num % denom
  qT = num // denom
  I = 0 if rT >= 0 else (1 if denom>0 else -1)
  qE = qT - I
  rE = rT + I * denom
  return qE 

def modE(num: int, denom: int) -> Tuple[int, int]:
  rT = num % denom
  qT = num // denom
  I = 0 if rT >= 0 else (1 if denom>0 else -1)
  qE = qT - I
  rE = rT + I * denom
  return rE 
```
